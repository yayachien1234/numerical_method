```python
import random
from copy import deepcopy


class Matrix:
# ----------建立陣列----------
    def __init__(self, nrows, ncols):
        """Construct a (nrows X ncols) matrix"""

        self.ncols = ncols
        self.nrows = nrows
        self.matrix = [[random.randint(0,10) for _ in range(self.ncols)] for _ in range(self.nrows)]
        pass
     
# ----------ADD----------
    def add(self, m):
        """return a new Matrix object after summation"""
        # Check two matrix are in the same size
        result = deepcopy(self)
#         判斷是否可以相加後用迴圈相加
        if (self.ncols==m.ncols)&(self.nrows==m.nrows):
            for i in range(0,self.nrows):
                for j in range(0,self.ncols):
                    result.matrix[i][j] = result.matrix[i][j]+m.matrix[i][j]
            pass
            return result
            # add self.matrix with m.matrix
        else:
            result = None
            return result
            


    def sub(self, m):
        """return a new Matrix object after substraction"""
        # Check two matrix are in the same size
        
        # substraction self.matrix from m.matrix
        result = deepcopy(self)
        pass
#     判斷是否可以相加後用迴圈相加
        if (self.ncols==m.ncols)&(self.nrows==m.nrows):
            for i in range(0,self.nrows):
                for j in range(0,self.ncols):
                    result.matrix[i][j] = result.matrix[i][j]-m.matrix[i][j]
    #         print(result.matrix)
            return result
        else:
            result = None
            return result
    
    def mul(self, m):
        """return a new Matrix object after multiplication"""
        # Check if the nrows of m is the same as ncols of self
        
        # Multiply self.matrix and m.matrix
        result = Matrix(self.nrows, m.ncols)
        pass
#     利用三個迴圈來實現矩陣相乘的結果
        if (self.ncols==m.nrows):
            for i in range(0,self.nrows):
                for j in range(0,m.ncols):
                    result.matrix[i][j]=0
                    for k in range(0,self.ncols):
                        result.matrix[i][j]= result.matrix[i][j]+ self.matrix[i][k]*m.matrix[k][j]
    #         print(result.matrix)
            return result
        else:
            result = None
            return result
# ----------Tranpose----------    
    def transpose(self):
        """return a new Matrix object after transpose"""

        pass
        result = Matrix(self.ncols, self.nrows)
#         判斷可以transpose後將i,j互換
        if self.matrix is not None:
            for i in range(0,self.ncols):
                for j in range(0,self.nrows):
                    result.matrix[i][j]=self.matrix[j][i]

            return result
        else:
            return None
    
    def display(self):
        """Display the content in the matrix"""
        pass
#     利用迴圈輸出矩陣的模樣
        for i in range(self.nrows):
            print(end = '| ')
            for j in range(self.ncols):
                print(self.matrix[i][j], end = '  ')
            print(end = '|')
            print('\n')
            
```

```python
a_rows = int(input("Enter A matrix's rows:"))
a_cols = int(input("Enter A matrix's cols:"))
print("Matrix A({}, {}):".format(a_rows, a_cols))
A = Matrix(a_rows, a_cols)
A.display()

b_rows = int(input("Enter B matrix's rows:"))
b_cols = int(input("Enter B matrix's cols:"))
print("Matrix B({}, {}):".format(b_rows, b_cols))
B = Matrix(b_rows, b_cols)
B.display()

print("="*10, 'A + B', "="*10)
result = A.add(B)
if result is not None:
    result.display()
else:
    print("Matrixs' size should be in the same size")

print("="*10, 'A - B', "="*10)
result = A.sub(B)
if result is not None:
    result.display()
else:
    print("Matrixs' size should be in the same size")

print("="*10, 'A * B', "="*10)
result = A.mul(B)
if result is not None:
    result.display()
else:
    print("Cannot mutiply matrix")
    
print("="*5, "the transpose of A*B", "="*5)
if result is not None:
    result = result.transpose()
    result.display()
else:
    print("Cannot transpose matrix")
```
